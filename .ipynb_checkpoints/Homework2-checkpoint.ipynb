##### 1.(a).1.1 Method 1: Removing All Missing Values

Since there are ```NaN``` values in the data, I will try removing rows with these values first.

````julia
# preprocessing: removing rows with NaN values
counter_NaN = 0
for i in 1: size(data)[1]
    if isnan(data[i, 2])
        counter_NaN = counter_NaN + 1
    end
end

purged_data = Array{Float64, 2}(undef, size(data)[1]-counter_NaN, size(data)[2])
current_NaN = 0
for i in 1: size(data)[1]
    if !isnan(data[i, 2])
        purged_data[i-current_NaN,:] = data[i,:]
    else
        current_NaN = current_NaN + 1
    end
end

days = purged_data[:, 1]
brightness = purged_data[:, 2]
plt.plot(1:(length(days)-1), [days[i+1]-days[i] for i in 1:(length(days)-1)])
plt.xlabel(L"$i$-th Observation")
plt.ylabel(L"$Day_{i+1}-Day_i$");
````

Removing missing values is not a good idea, because the plot above shows that observations are no longer uniformly distributed, thus causing trouble in find the period of the orbital period.  

##### 1.(a).1.2 Method 2: Linear Interpolation

````julia
purged_data = data[:,:] # copy the data instead

for i in 1: size(purged_data)[1]
    if isnan(purged_data[i, 2])
        left = right = i # find the indices of the left and right nearest values that are not NaN
        for j in i: -1: 1
            if !isnan(purged_data[j, 2])
                left = j
                break
            end
        end
        for j in i: +1: size(purged_data)[1]
            if !isnan(purged_data[j, 2])
                right = j
                break
            end
        end
        weight_l = (purged_data[i,1]-purged_data[left,1]) / (purged_data[right,1]-purged_data[left,1])
        weight_r = (purged_data[right,1]-purged_data[i,1]) / (purged_data[right,1]-purged_data[left,1])
        # use convex combination of these two NaN values to interpolate the missing value
        purged_data[i, 2] = weight_l * purged_data[left,2] + weight_r * purged_data[right,2] 
    end
end


days = purged_data[:, 1]
brightness = purged_data[:, 2]
plt.plot(1:(length(days)-1), [days[i+1]-days[i] for i in 1:(length(days)-1)])
plt.ylim(0, 5*10^(-2))
plt.xlabel(L"$i$-th Observation")
plt.ylabel(L"$Day_{i+1}-Day_i$");
````

Linear interpolation preserves the distribution of the data, which means that in the processed data, observations are still made based on an interval of around $0.02$ days.

#### 1.(a).3 Find the Period

From the plot above, there seems to be a periodic feature in the data. Suppose the model is

$$
B(t) =  B(t-m) + \epsilon \tag{1}
$$

where $t$ denotes the time, $B(t)$ denotes the brightness of the star at time $t$, $m$ is the orbital period of the planet, and $\epsilon$ is the noise term.


We have read the data into two arrays ```brightness``` and ```days```, so we can use array indices to represent $t$. For example, $t = 120.53881583872862$, which is the first element of ```days```, will be denoted as ```days[1]``` (or $D_1$ in the math formulas), and the corresponding brightness, $1034944.9375$, will be denoted as ```brightness[1]``` (or $B_1$).

Suppose the time period is $m$, which means for any arbitrary $i$ such that the following indices make sense,
```brightness[i]``` observed at ```days[i]``` should be approximately equal to ```brightness[i-m]``` observed at ```days[i-m]```.

Let the sum of square errors be

$$
\begin{equation}\label{eqn1}
\sum_{1 \le i,\, i+m \le n} (B_{i} - B_{i+m})^2, \tag{2}
\end{equation}
$$

where $n$ is ```size(purged_data)[1]```.

The objective is to minimise the mean of the sum of squares. 

````julia
"""
This function calcluate the mean squared error, where the sum of squares is defined by (2).

Inputs:
- data: the brightness of the star. (Array::Float64)
- m: the period of the planet. (Int)

Outputs:
- mean_error: the mean of the sum of squared errors.
"""
function mean_error(data, m)
    counter_example = 1 # this counts how many examples can be used in calculating the error.
    n = length(data)
    sum_error = 0
    while counter_example + m < n
        sum_error = sum_error + (data[counter_example] - data[counter_example+m])^2
        counter_example = counter_example + 1
    end
    mean_error = sum_error / counter_example
    return mean_error
end
````

````julia
n = size(purged_data)[1]

ms = 1: Int(round(n/2)) # possilbe periods
mean_errors = Array{Float64, 1}(undef, length(ms));

for m in ms
    mean_errors[m] = mean_error(brightness, m)
end

(min_mean_error, m) = findmin(mean_errors); # find the minimum mean squared error.
````

Now we get the period in the form of indices, but we need to transform it into days.

````julia
period = mean([days[i+m]-days[i] for i in 1: n-m])
@show period;
````

So the orbital period is around $2.2069083435310897$ days.

### Part (b)

#### 1.(b).1 Data Analysis and Assumption Proposal

A dip corresponds to a trough in the plot of data (shown below).

````julia
@show plot_data
````

So when brightness reaches the lowest point in a trough, the centers of the planet and the star coincide.

Let $R$ denote the radius of the star, $r$ the radius of the planet, $c_s$ the center of the star, $c_p$ the center of the planet.

Assume the relation of $B$ (the brightness of the star) $r$ and $R$ is simply

$$
B = C \times [R^2 - \delta(c_s,c_p)r^2]
$$

where $C \in \mathbb{R}$ is a constant and $\delta(c_s,c_p)$ indicates whether the centers coincide or not.

When there is no dipping, the model becomes

$$
B = C \times R^2.
$$

Since both $C$ and $R$ are unknown, we cannot use multinomial regression to solve $C$. Instead, let us treat $ C \times R^2 $ as one entity, and we can use the mean of its observations $B_i$ to estimate it.

But the tricky part is how to differentiate between dips and non-dips. In the general, the brightness declines, so we have to use different thresholds to distinguish. For simplicity, I assume

$$
\begin{cases}
B > \text{threshold 1 } \& \text{ no dips have occured} \implies & \text{non-dip}\\
B = \text{threshold 1 } \& \text{ no dips have occured} \implies & \text{1st dip begins}\\
B = \text{threshold 1 } \& \text{ 1st dip has begun but not ended} \implies & \text{1st ends}\\
B > \text{threshold 2 } \& \text{ 1st dip has ended, but 2nd dip has not begun} \implies & \text{non-dip}\\
B = \text{threshold 2 } \& \text{ 1st dip has ended, but 2nd dip has not begun} \implies & \text{2nd dip begins}\\
B = \text{threshold 2 } \& \text{ 2nd dip has begun but not ended} \implies & \text{2nd dip ends} \\
B > \text{threshold 3 } \& \text{ 2nd dip has ended, but 3rd dip has not begun} \implies & \text{non-dip} \\
\vdots & \vdots
\end{cases}
$$

````julia
# set thresholds
threshold1 = 1.0349*10^6 
threshold2 = 1.0349*10^6 
threshold3 = 1.0347*10^6 
threshold4 = 1.0344*10^6 
threshold5 = 1.0341*10^6


dip_begin_time = Array{Float64 ,1}(undef, 5)
dip_begin_ind = Array{Int64 ,1}(undef, 5)
dip_end_time = Array{Float64 ,1}(undef, 4)
dip_end_ind = Array{Int64 ,1}(undef, 4)


counter = 1
label = true
while counter  length(brightness)
    if label && brightness[counter]  threshold1
        dip_begin_time[1] = days[counter]
        dip_begin_ind[1] = counter
        label = false
    elseif !label && brightness[counter]  threshold1
        dip_end_time[1] = days[counter]
        dip_end_ind[1] = counter
        label = true
        break
    end
    counter = counter + 1
end

while counter  length(brightness)
    if label && brightness[counter]  threshold2
        dip_begin_time[2] = days[counter]
        dip_begin_ind[2] = counter
        label = false
    elseif !label && brightness[counter]  threshold2
        dip_end_time[2] = days[counter]
        dip_end_ind[2] = counter
        label = true
        break
    end
    counter = counter + 1
end

while counter  length(brightness)
    if label && brightness[counter]  threshold3
        dip_begin_time[3] = days[counter]
        dip_begin_ind[3] = counter
        label = false
    elseif !label && brightness[counter]  threshold3
        dip_end_time[3] = days[counter]
        dip_end_ind[3] = counter
        label = true
        break
    end
    counter = counter + 1
end

while counter  length(brightness)
    if label && brightness[counter]  threshold4
        dip_begin_time[4] = days[counter]
        dip_begin_ind[4] = counter
        label = false
    elseif !label && brightness[counter]  threshold4
        dip_end_time[4] = days[counter]
        dip_end_ind[4] = counter
        label = true
        break
    end
    counter = counter + 1
end

while counter  length(brightness)
    if label && brightness[counter]  threshold5
        dip_begin_time[5] = days[counter]
        dip_begin_ind[5] = counter
        label = false
        break
    end
    counter = counter + 1
end

nondip1 = days[1]: 0.01: dip_begin_time[1]
nondip2 = dip_end_time[1]: 0.01: dip_begin_time[2]
nondip3 = dip_end_time[2]: 0.01: dip_begin_time[3]
nondip4 = dip_end_time[3]: 0.01: dip_begin_time[4]
nondip5 = dip_end_time[4]: 0.01: dip_begin_time[5]

plt.figure(figsize=(10, 4))
plt.plot(days, brightness, label="data")
plt.plot(nondip1, threshold1 * ones(length(nondip1)), ":", label="threshold 1")
plt.plot(nondip2, threshold2 * ones(length(nondip2)), ":", label="threshold 2")
plt.plot(nondip3, threshold3 * ones(length(nondip3)), ":", label="threshold 3")
plt.plot(nondip4, threshold4 * ones(length(nondip4)), ":", label="threshold 4")
plt.plot(nondip5, threshold5 * ones(length(nondip5)), ":", label="threshold 5")
plt.plot(days[1]*ones(length(1.027*10^6:threshold1)), 1.027*10^6:threshold1, "k--")
plt.plot(dip_begin_time[1]*ones(length(1.027*10^6:threshold1)), 1.027*10^6:threshold1, "k--")
plt.plot(dip_end_time[1]*ones(length(1.027*10^6:threshold1)), 1.027*10^6:threshold1, "k--")
plt.plot(dip_begin_time[2]*ones(length(1.027*10^6:threshold2)), 1.027*10^6:threshold2, "k--")
plt.plot(dip_end_time[2]*ones(length(1.027*10^6:threshold2)), 1.027*10^6:threshold2, "k--")
plt.plot(dip_begin_time[3]*ones(length(1.027*10^6:threshold3)), 1.027*10^6:threshold3, "k--")
plt.plot(dip_end_time[3]*ones(length(1.027*10^6:threshold3)), 1.027*10^6:threshold3, "k--")
plt.plot(dip_begin_time[4]*ones(length(1.027*10^6:threshold4)), 1.027*10^6:threshold4, "k--")
plt.plot(dip_end_time[4]*ones(length(1.027*10^6:threshold4)), 1.027*10^6:threshold4, "k--")
plt.plot(dip_begin_time[5]*ones(length(1.027*10^6:threshold5)), 1.027*10^6:threshold5, "k--")
plt.xlim(120, 131)
plt.ylim(1.027*10^6, 1.036*10^6)
plt.xlabel("days")
plt.ylabel("brightness")
plt.legend()
plt.title("Brightness versus Days");
````

````julia
nondip1_brightness = brightness[1: (dip_begin_ind[1]-1)]
nondip2_brightness = brightness[(dip_end_ind[1]+1): (dip_begin_ind[2]-1)]
nondip3_brightness = brightness[(dip_end_ind[2]+1): (dip_begin_ind[3]-1)]
nondip4_brightness = brightness[(dip_end_ind[3]+1): (dip_begin_ind[4]-1)]
nondip_brightness = vcat(nondip1_brightness,nondip2_brightness,
                         nondip3_brightness,nondip4_brightness)

CR = mean(nondip_brightness)
@show CR;
````

So

$$C \times R^2 \approx 1.0349372596590122\times10^6.$$

There are five troughs, so use the mean of the minima of brightness in these troughs to estimate 

$$C \times (R^2 - r^2).$$

````julia
lowest_points = Array{Float64, 1}(undef, 5)

for i in 1: length(brightness)
    if dip_begin_ind[1]  i  dip_end_ind[1]
        if brightness[i-1] > brightness[i] < brightness[i+1]
            lowest_points[1] = brightness[i]
        end
    end
    if dip_begin_ind[2]  i  dip_end_ind[2]
        if brightness[i-1] > brightness[i] < brightness[i+1]
            lowest_points[2] = brightness[i]
        end
    end
    if dip_begin_ind[3]  i  dip_end_ind[3]
        if brightness[i-1] > brightness[i] < brightness[i+1]
            lowest_points[3] = brightness[i]
        end
    end
    if dip_begin_ind[4]  i  dip_end_ind[4]
        if brightness[i-1] > brightness[i] < brightness[i+1]
            lowest_points[4] = brightness[i]
        end
    end
    if dip_begin_ind[5]  i  (length(brightness)-1)
        if brightness[i-1] > brightness[i] < brightness[i+1]
            lowest_points[5] = brightness[i]
        end
    end
end

CR_r = mean(lowest_points)
@show CR_r;
````

So an estimation of $C\times(R^2 - r^2)$ is $1.0278559375\times10^6$, and thus

$$
\begin{align*}
C \times r^2 = & C\times R^2 - C\times (R^2 - r^2) \\
\approx & 1.0349372596590122 \times 10^6 - 1.0278559375\times10^6 \\
= & 0.007081322159012249 \times 10^6,
\end{align*}
$$

and 

$$
\begin{align*}
\frac{r}{R} = & \sqrt{\frac{Cr^2}{CR^2}} \\ 
\approx & \sqrt{\frac{0.007081322159012249 \times 10^6}{1.0349372596590122 \times 10^6}} \\ 
\approx & 0.08271802662059785.
\end{align*}
$$

## Q5.3.2 Auto-covariance of an Autoregressive Model

* Simulate a first order autoregressive model with your parameter choice  
for $c$, $\phi$ and $\sigma_\epsilon$ (but with $|\phi|<1$).


* Run the simulation for $5000$ steps and demonstrate that $<{X}>$ and Var$(X)$ are comparable to theory.



* Extract the auto-covariance from the data.  
Provide a plot comparing it to the theoretical prediction.

Assume the autoregressive model is 

$$
X_t = 10 + 0.99 X_{t-1} + \epsilon_t,
$$

with $\epsilon_t \sim \mathcal{N}(0, 1)$.

Let $X_0$ be chosen by the theoretically mean $\frac{10}{1-0.99} = 1000$.

````julia
steps = 5000
Xs = Array{Float64, 1}(undef, steps+1) # Xs[1] = X_0
Xs[1] = 1000
for i in 1: steps
    Xs[i+1] = 10 + 0.99*Xs[i] + randn()
end

plt.figure(figsize=(12, 5))
plt.plot(1:length(Xs), Xs, ":")
plt.xlabel(L"$t$")
plt.ylabel(L"$X_t$")
plt.title(L"$X_t = 1 + 0.99 X_{t-1} + \epsilon_t$");
````

````julia
# sample mean
sample_mean = mean(Xs[begin+1: end])
# sample variance
sample_var = var(Xs[begin+1: end])

@show sample_mean
@show sample_var;
````

Taking expectation of the both sides of 

$$X_t = 10 + 0.99 X_{t-1} + \epsilon_t$$

results in 

$$\mathbb{E}[X] = 10 + 0.99 \mathbb{E}[X],$$

so theoretically

$$\mathbb{E}[X] = \frac{10}{0.01} = 1000,$$ 

which is very close to the sample mean $1001.2401695434858$.

Similarly, taking variance of the both sides gives

$$\text{Var}[X] = 0.99^2 \text{Var}[X] + \text{Var}[\epsilon_t] = 0.9801 \text{Var}[X] + 1,$$

so the theoretical variance is

$$\text{Var}[X] = \frac{1}{1-0.9801} \approx 50.25125628140703,$$

which is not very close to the sample variance $41.69057336713657$.

By 

$$X_t = 10 + 0.99 X_{t-1} + \epsilon_t,$$

we have 

$$
\begin{align*}
X_t = & 10 + 0.99 (10 + 0.99X_{t-2} + \epsilon_{t-1}) +\epsilon_t \\
= & 10(0.99^0 + 0.99^1) + 0.99^2 X_{t-2} + 0.99 \epsilon_{t-1} + 0.99^0\epsilon_t.
\end{align*}
$$

So by induction, we have 

$$
\begin{align*}
X_t = & 10(0.99^0 + \cdots + 0.99^{t-1}) + 0.99^t X_0 + 0.99^{t-1} \epsilon_1 + 0.99^{t-2} \epsilon_2 + \cdots + 0.99^0 \epsilon_t \\
= & 10 \times \frac{1-0.99^t}{1-0.99} + 0.99^t X_0 + \sum_{k = 1}^t 0.99^{t-k} \epsilon_k \\ 
= & 1000 \times (1 - 0.99^t) + 0.99^t X_0 + \sum_{k = 1}^t 0.99^{t-k} \epsilon_k \\ 
= & 1000 \times (1 - 0.99^t) + 0.99^t \times 1000 + \sum_{k = 1}^t 0.99^{t-k} \epsilon_k \\
= & 1000 + \sum_{k = 1}^t 0.99^{t-k} \epsilon_k
\end{align*}
$$

So $\forall n \in \mathbb{N}_+$, 

$$
X_{t+n} = 1000  + \sum_{k=1}^{t+n} 0.99^{t+n-k} \epsilon_k.
$$

As a result

$$
\begin{align*}
\gamma_t(n) = & \mathbb{E}\left[(X_t - \mathbb{E}[X_t]) (X_{t+n} - \mathbb{E}[X_{t+n}]) \right] \\ 
= & \mathbb{E}\left[ ( \sum_{k=1}^t 0.99^{t-k} \epsilon_k ) ( \sum_{k=1}^{t+n} 0.99^{t+n-k} \epsilon_k ) \right] \\
= & \mathbb{E}\left[ ( \sum_{k=1}^t 0.99^{t-k} \epsilon_k ) (0.99^n \sum_{k=1}^{t+n} 0.99^{t-k} \epsilon_k ) \right] \\ 
= & 0.99^n \mathbb{E}\left[ ( \sum_{k=1}^t 0.99^{t-k} \epsilon_k ) ( \sum_{k=1}^{t+n} 0.99^{t-k} \epsilon_k ) \right] \\
= & 0.99^n \sum_{k=1}^t 0.99^{2(t-k)} \sigma_\epsilon^2 \\
= & 0.99^n \cdot \sigma_\epsilon^2 \cdot \frac{1 \cdot (1-0.99^{2t})}{1- 0.99^2} \\
= & 0.99^n \cdot 1 \cdot \frac{1 \cdot (1-0.99^{2t})}{1-0.9801} \\
= & 0.99^n \cdot \frac{1-0.99^{2t}}{0.0199}.
\end{align*}
$$


Suppose the process has been running for a long time, so let $t \to \infty$ and we get 

$$
\gamma(n) = \lim_{t\to\infty} \gamma_t(n) = \frac{0.99^n}{0.0199}.
$$

````julia
ns = 0: (steps-1)

# theoretical autocovariance
theory_autocov(n) = 0.99.^n ./ 0.0199
theory_autocovs = theory_autocov(ns)

# sample autocovariance
sample_autocovs = autocov(Xs[2:end], ns)


ns = -ns[end]: 1: ns[end]
theory_autocovs = vcat(reverse(theory_autocovs)[1: (end-1)], theory_autocovs)
sample_autocovs = vcat(reverse(sample_autocovs)[1: (end-1)], sample_autocovs)
plt.figure(figsize=(12,5))
plt.plot(ns, theory_autocovs, label=L"theoretical auto-covariance: $\frac{0.99^n}{0.0199}$")
plt.plot(ns, sample_autocovs, 
    label=L"sample auto-covariance: $\frac{1}{m}\sum_{t=1}^{m-n}(x_{t+n} -\bar{x})(x_t - \bar{x})$")
plt.legend()
plt.xlim(-steps, steps)
plt.xlabel(L"$n$")
plt.ylabel("autocovariance")
plt.title("Theoretical Autocovariance vs Sample Autocovariance");
````

## Q5.3.3 Distribution of Sample Means for an Autoregressive Process 

*  Simulate a first-order autoregressive process for $100,000$ steps with $c=0$, $\phi=0.9$ and $\sigma_\epsilon=2$.  


*  Calculate the distribution of sample means in two ways. For the first way, take $10,000$ random samples of $10$ adjacent data points. For the second way, take $10,000$ random samples of $10$ data points each picked from any step of your autoregressive process.  


* Plot out the histograms of the sample means for these two cases and examine if they agree with what would be expected from a naive application of the central-limit theorem (you can assume knowledge of the true mean and variance used to generate the autoregressive process). Comment on the differences if any between the two ways of calculating the sample means.  

````julia
c = 0
 = 0.9 
 = 2
steps = 10^5

Xs = Array{Float64, 1}(undef, steps+1)
Xs[1] = 0

for i in 1: steps
    Xs[i+1] = c + *Xs[i] + *randn()
end
````

The mean of the process is 

$$\mathbb{E}[X] = \frac{c}{1-\phi} = 0$$

and the variance is

$$\text{Var}[X] = \frac{\sigma_\epsilon^2}{1 - \phi^2} = \frac{4}{1-0.81} = \frac{4}{0.19}.$$

So if these $X_t$'s are i.i.d. distributed with mean $\mu =0$ and variance $\sigma^2 = \frac{4}{0.19}$, by the [Central Limit Theorem](https://en.wikipedia.org/wiki/Central_limit_theorem), we have 

$$\sqrt{n} (\bar{X}_n - \mu) \xrightarrow{\text{d}} \mathcal{N}(0, \sigma^2).$$

So we can expect the sample mean with $10$ samples $\bar{X}_{10}$ to ba approximately normally distributed with $\mu = 0$ and $\sigma^2 = \frac{40}{19}$, namely

$$\bar{X}_{10} \sim \mathcal{N}(0, \frac{40}{19}).$$

### Way 1

````julia
sample_means1 = Array{Float64, 1}(undef, 10^4)
for i in 1: 10^4
    start_ind = rand(1: (length(Xs)-9))
    random_sample = Xs[start_ind: (start_ind+9)]
    sample_means1[i] = mean(random_sample)
end
````

````julia
nbins = 50
plt.hist(sample_means1, bins=nbins, density=true, label=L"$\bar{X}_{10}$")

xrange = -15: 0.01: 15
pdf_vals = pdf(Normal(0, sqrt(40/19)), xrange)
plt.plot(xrange, pdf_vals, 
    label=L"$\frac{1}{\sqrt{80\pi/19}} \exp \left(\frac{19x^2}{80}\right)$")
plt.legend()
plt.xlim(-15, 15)
plt.xlabel(L"$x$")
plt.ylabel(L"$y$")
plt.title(L"Sample Mean ($\bar{X}_{10}$) Distribution vs $\mathcal{N}(0, 40/19)$");
````

If random samples are composed of $10$ adjacent data points, then the sample mean seems not follow $\mathcal{N}(0, \frac{40}{19})$.

````julia
lags = 0: 30
auto_correlations = autocor(Xs, lags)
lags = -lags[end]: 1: lags[end]
auto_correlations = vcat(reverse(auto_correlations)[1: (end-1)], auto_correlations)


plt.plot(lags, auto_correlations)
plt.xlabel("lags")
plt.ylabel("auto-correlation")
plt.title(L"Auto-Correlation of $X_t$ vs lags");
````

From the plot above, we can see $10$ consecutive data points still have very strong correlations, so the i.i.d. assumption in the Central Limit Theorem does not hold. Thus, using the Central Limit Theorem to estimate the distribution of the sample mean $\bar{X}_{10}$ is a bad idea.

#### Way 2

If we simply randomly select $10$ data points and use them to compose a sample, then the sample mean $\bar{X}_{10}$ coincides with $\mathcal{N}(0, 40/19)$ very well. Because we chose the data randomly, these points are much less dependent on each other. Thus, the i.i.d. assumption is well satisfied, and we can use the Central Limit Theorem to approximate the distribution of the sample mean $\bar{X}_{10}$.

## Q6.3.1 Power-Spectrum for the O-U Process

* Consider the Ornstein-Uhlenbeck process described in the lecture notes.   
Use it's auto-covariance to provide the form of the power spectrum $S(\omega)$.  

Plug 

$$ A(t) = \sigma^2 e^{-\frac{|t|}{\tau}}$$

into 

$$ S(\omega) = \frac{1}{T} \int_{-\frac{T}{2}}^\frac{T}{2} \left( \int_{-\infty}^\infty e^{-i\omega t} A(t) d t \right) dr,$$

then we have 

$$
\begin{align*}
S(\omega) = &\frac{1}{T} \int_{-\frac{T}{2}}^\frac{T}{2} \left( \int_{-\infty}^\infty e^{-i\omega t} \sigma^2 e^{-\frac{|t|}{\tau}} d t \right) dr \\
= & \frac{1}{T} \cdot T \cdot \sigma^2 \int_{-\infty}^\infty e^{-i\omega t} e^{-\frac{|t|}{\tau}} d t \\
= & \sigma^2 \int_{-\infty}^\infty e^{-\frac{|t|}{\tau}} [\cos(\omega t) - i \sin(\omega t)] dt \\ 
= & \sigma^2 \int_{-\infty}^\infty e^{-\frac{|t|}{\tau}} \cos(\omega t) dt \\
= & 2 \sigma^2 \int_{0}^\infty e^{-\frac{t}{\tau}} \cos(\omega t) dt.
\end{align*}
$$

By 

$$
\begin{align*}
\int_0^\infty e^{-\frac{t}{\tau}} \cos(\omega t) dt = & -\tau \int_0^\infty \cos(\omega t) d e^{-\frac{t}{\tau}} \\
= & - \tau \left( e^{-\frac{t}{\tau}} \cos (\omega t) |_0^\infty - \int_0^\infty e^{-\frac{t}{\tau}} d\cos(\omega t) \right) \\
= & - \tau \left( -1 + \omega \int_0^\infty e^{-\frac{t}{\tau}}\sin(\omega t) dt \right) \\
= & \tau - \tau \omega \int_0^\infty e^{-\frac{t}{\tau}}\sin(\omega t) dt \\
= & \tau + \tau^2 \omega \int_0^\infty \sin(\omega t) de^{-\frac{t}{\tau}} \\
= & \tau + \tau^2 \omega \left( e^{-\frac{t}{\tau}}\sin(\omega t)|_0^\infty - \int_0^\infty e^{-\frac{t}{\tau}}d \sin(\omega t) \right) \\
= & \tau + \tau^2 \omega \left( 0 - \omega \int_0^\infty e^{-\frac{t}{\tau}} \cos(\omega t)d t \right)\\
= & \tau - \tau^2 \omega^2 \int_0^\infty e^{-\frac{t}{\tau}} \cos(\omega t)d t,
\end{align*}
$$

we have 

$$
\int_0^\infty e^{-\frac{t}{\tau}} \cos(\omega t)d t = \frac{\tau}{1+\tau^2\omega^2}.
$$

So 

$$
S(\omega) = \frac{2\tau\sigma^2}{1 + \tau^2\omega^2}.
$$

## Q6.3.2 Correlated O-U Processes

* Consider two O-U processes driven by the same Gaussian white noise $\xi(t)$  
But with two different time constants $\tau_x$ and $\tau_y$ so that

$\hspace{0.8cm} \tau_x\frac{dx}{dt}=-x+\sigma\sqrt{2\tau_x}\xi(t)~$ and $\tau_y\frac{dy}{dt}=-y+\sigma\sqrt{2\tau_y}\xi(t)~$


* What is the same-time correlation $\langle{x(t)y(t)}\rangle$?


* Calculate the cross-covariance $\langle{x(t)y(t+T)}\rangle$ and be sure to   
provide the forms for both positive and negative $T$.

From 

$$\tau_x\frac{dx}{dt}=-x+\sigma\sqrt{2\tau_x}\xi(t)$$

and 

$$\tau_y\frac{dy}{dt}=-y+\sigma\sqrt{2\tau_y}\xi(t)~,$$

we have 

$$
X(t) = \sigma \sqrt{2\tau_x} \int_{-\infty}^t e^{-\frac{t-s}{\tau_x}} \xi(s) \frac{ds}{\tau_x}
$$

and

$$
Y(t) = \sigma \sqrt{2\tau_y} \int_{-\infty}^t e^{-\frac{t-s}{\tau_y}} \xi(s) \frac{ds}{\tau_y}
$$

So 

$$
\begin{align*}
\mathbb{E} [X(t)Y(t)] = & 2\sigma^2 \sqrt{\tau_x \tau_y} \int_{-\infty}^t \left( \int_{-\infty}^t e^{- \frac{t-s}{\tau_x} - \frac{t-s'}{\tau_y}} \mathbb{E}[\xi(s)\xi(s')] \frac{d s'}{\tau_y} \right) \frac{ds}{\tau_x} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} \int_{-\infty}^t \left( \int_{-\infty}^t e^{- \frac{t-s}{\tau_x} - \frac{t-s'}{\tau_y}} \delta(s-s') ds' \right) ds \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} \int_{-\infty}^t e^{-\frac{t-s}{\tau_x} - \frac{t-s}{\tau_y}} ds \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} \int_{-\infty}^t e^{\frac{ - (\tau_x +\tau_y) t + (\tau_x + \tau_y) s}{\tau_x\tau_y}} ds \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{(\tau_x +\tau_y) t }{\tau_x\tau_y}} \int_{-\infty}^t e^{\frac{(\tau_x + \tau_y) s}{\tau_x\tau_y}} ds \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{(\tau_x +\tau_y) t }{\tau_x\tau_y}} \frac{\tau_x\tau_y}{\tau_x+\tau_y}\int_{-\infty}^t d e^{\frac{(\tau_x + \tau_y) s}{\tau_x\tau_y}} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{(\tau_x +\tau_y) t }{\tau_x\tau_y}} \frac{\tau_x\tau_y}{\tau_x+\tau_y} \left( \left. e^{\frac{(\tau_x + \tau_y) s}{\tau_x\tau_y}} \right|_{-\infty}^t \right) \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{(\tau_x +\tau_y) t }{\tau_x\tau_y}} \frac{\tau_x\tau_y}{\tau_x+\tau_y} e^{\frac{(\tau_x + \tau_y) t}{\tau_x\tau_y}} \\
= & \frac{2 \sigma^2 \sqrt{\tau_x\tau_y}}{\tau_x+\tau_y}.
\end{align*}
$$

As for $\mathbb{E}[X(t)Y(t+T)]$, we have calculated the case when $T = 0$. Now, assume $T > 0$.

By 

$$
X(t) = \sigma \sqrt{2\tau_x} \int_{-\infty}^t e^{-\frac{t-s}{\tau_x}} \xi(s) \frac{ds}{\tau_x}
$$

and 

$$
Y(t+T) = \sigma \sqrt{2\tau_y} \int_{-\infty}^{t+T} e^{-\frac{t+T-s}{\tau_y}} \xi(s) \frac{ds}{\tau_y},
$$

we have 

$$
\begin{align*}
\mathbb{E} [X(t) Y(t+T)] = & 2 \sigma^2 \sqrt{\tau_x\tau_y} \int_{-\infty}^t \left( \int_{-\infty}^{t+T} e^{- \frac{t-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \mathbb{E}[\xi(s) \xi(s')] \frac{ds'}{\tau_x} \right) \frac{ds}{\tau_y} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} \int_{-\infty}^t \left( \int_{-\infty}^{t+T} e^{- \frac{t-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{T}{\tau_y}} \int_{-\infty}^t \left( \int_{-\infty}^{t+T} e^{- \frac{t-s}{\tau_x} - \frac{t-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{T}{\tau_y}} \int_{-\infty}^t \left( \int_{-\infty}^{t} e^{- \frac{t-s}{\tau_x} - \frac{t-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds}  \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{-\frac{T}{\tau_y}} \int_{-\infty}^t e^{- \frac{t-s}{\tau_x} - \frac{t-s}{\tau_y}} ds \\
= & \frac{2 \sigma^2 \sqrt{\tau_x\tau_y}}{\tau_x+\tau_y} e^{-\frac{T}{\tau_y}}.
\end{align*}
$$

Similarly, when $T < 0$, we have 

$$
\begin{align*}
\mathbb{E} [X(t) Y(t+T)] = & 2 \sigma^2 \sqrt{\tau_x\tau_y} \int_{-\infty}^t \left( \int_{-\infty}^{t+T} e^{- \frac{t-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \mathbb{E}[\xi(s) \xi(s')] \frac{ds'}{\tau_x} \right) \frac{ds}{\tau_y} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} \int_{-\infty}^t \left( \int_{-\infty}^{t+T} e^{- \frac{t-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{\frac{T}{\tau_x}} \int_{-\infty}^{t} \left( \int_{-\infty}^{t+T} e^{- \frac{t+T-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds} \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{\frac{T}{\tau_x}} \int_{-\infty}^{t+T} \left( \int_{-\infty}^{t+T} e^{- \frac{t+T-s}{\tau_x} - \frac{t+T-s'}{\tau_y}} \delta(s-s') {ds'}\right) {ds}  \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{\frac{T}{\tau_x}} \int_{-\infty}^{t+T} e^{- \frac{t+T-s}{\tau_x} - \frac{t+T-s}{\tau_y}} ds \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{\frac{T}{\tau_x}} \int_{-\infty}^{t+T} e^{- \frac{t-(s-T)}{\tau_x} - \frac{t-(s-T)}{\tau_y}} d(s-T) \\
= & \frac{2\sigma^2}{\sqrt{\tau_x\tau_y}} e^{\frac{T}{\tau_x}} \int_{-\infty}^{t} e^{- \frac{t-s}{\tau_x} - \frac{t-s}{\tau_y}} ds \\
= & \frac{2 \sigma^2 \sqrt{\tau_x\tau_y}}{\tau_x+\tau_y} e^{\frac{T}{\tau_x}}.
\end{align*}
$$

So in conclusion, we have 

$$
\mathbb{E} [X(t) Y(t+T)] = 
\begin{cases}
2 \sigma^2 \frac{\sqrt{\tau_x\tau_y}}{\tau_x+\tau_y} e^{\frac{T}{\tau_x}} \quad & T < 0 \\
2 \sigma^2 \frac{\sqrt{\tau_x\tau_y}}{\tau_x+\tau_y} \quad & T = 0 \\
2 \sigma^2 \frac{\sqrt{\tau_x\tau_y}}{\tau_x+\tau_y} e^{-\frac{T}{\tau_y}} \quad & T > 0
\end{cases}
$$
