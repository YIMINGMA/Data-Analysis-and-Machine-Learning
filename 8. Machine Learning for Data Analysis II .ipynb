### 8.1.1 Network with Hidden Layer  $\renewcommand\p{\partial}$

Generalise input-output network of Day 7 to network with hidden layer.   


* The input matrix $\mathbf{X}$ (or $\mathbf{\tilde{X}}$ with the bias) and target vector $\mathbf{T}$ have same meaning as before, as does $\mathbf{P}$ the prediction vector.


* The hidden layer has $n_h$ units as well as a bias. The matrix of its values for all samples is $\mathbf{H}$ with dimension $(n_s,n_h)$. An additional matrix $\mathbf{\tilde{H}}$ is defined that includes the bias as a final column of ones, and has dimension $(n_s,n_h+1)$. 


* The matrix of weights $\mathbf{w}$ between layers $\tilde{X}$ and $H$ has dimensions $(n_x+1,n_h)$
* The matrix of weights $\mathbf{v}$ between layers $\tilde{H}$ and $P$ has dimensions $(n_h+1,1)$.


* Use sigmoid non-linearities of the linear input $z$ so that $f(z)=1/(1+\exp(-z))$.  


* The prediction of the network can therefore be written in matrix form   



$$
f(\tilde{\mathbf{X}}\mathbf{w})=\mathbf{H}~~\mbox{ and} ~~f(\tilde{\mathbf{H}}\mathbf{v})=\mathbf{P}.
$$


* We will use the cost function as before  

$$C=-\frac{1}{n_s}\sum_s\left[T_s\log(P_s)+(1-T_s)\log(1-P_s)\right].$$

## 8.1.2 Gradient Descent for Weights $v$  

* Now have two sets of weights $w$ and $v$.  
We need to calculate forms for the cost-function gradient in respect to $w_{ij}$ and $v_j$.   


* The network is defined by the formulae


$$
f(\tilde{\mathbf{X}}\mathbf{w})=\mathbf{H}~~\mbox{ and} ~~f(\tilde{\mathbf{H}}\mathbf{v})=\mathbf{P} ~~\mbox{ and }~~C=-\frac{1}{n_s}\sum_s\left[T_s\log(P_s)+(1-T_s)\log(1-P_s)\right].
$$



* **<font color='red'>Question. </font>** For a single data point where $c=-t\log(p)-(1-t)\log(1-p)$ show that   


$$ \frac{dc}{dv_j}=\tilde{h}_j\delta^p~~~\mbox{ so that in matrix form}~~~ 
\frac{dC}{d\mathbf{v}}=\frac{1}{n_s}\tilde{\mathbf{H}}'\Delta^p
$$

$\hspace{0.6cm}$ where $\delta^p=(p-t)$ and $\Delta^p_s=P_s-T_s$ is a matrix of dimension $(n_s,1)$. 



* **<font color='red'>Question. </font>** Using the single data-point cost function $c=-t\log(p)-(1-t)\log(1-p)$ show that    



$$
\frac{dc}{dw_{ij}}=\frac{dc}{dp}\frac{dp}{dz^p}\frac{dz^p}{dh_j}\frac{dh_j}{dw_{ij}}=\tilde{x}_i\delta^h_j ~~~\mbox{ where }~~~\delta_j^h=h_j(1-h_j)\delta^pv_j~~~\mbox{ and $j$ runs from 1 to $n_h$}
$$


* This gives the gradient for the $\mathbf{w}$ weights as
$$
\frac{dC}{d\mathbf{w}}=\frac{1}{n_s}\tilde{\mathbf{X}}'\Delta^h 
$$


$\hspace{0.6cm}$ where the $(n_s,n_h)$ dimensional matrix $\Delta^h$ has entries  


$$
\Delta^h_{sj}=H_{sj}(1-H_{sj})[\Delta_s^pv_j] ~~\mbox{ where }~~  \Delta^p_s=P_s-T_s
$$


* The following relations are all that is needed to teach the network


$$
\frac{dC}{d\mathbf{v}}=\frac{1}{n_s}\tilde{\mathbf{H}}'\Delta^p  ~~\mbox{ and }~~\frac{dC}{d\mathbf{w}}=\frac{1}{n_s}\tilde{\mathbf{X}}'\Delta^h
$$

### 8.2.1 Creating the Data 

* This time the data will not be separable by a straight line. 


* Consider two circles centred at (0.3,0.3) and (-0.5,0.5) both of radius 0.3.


* Generate $n_s=400$ points on the plane distributed as in the previous case.


* If the points fall within either circle they are assigned $t=1$, otherwise $t=0$.


* Generate the matricies $\mathbf{X}$ and $\mathbf{T}$ and plot out the scatter of data points.

```julia
ns=400

# two circular blobs - requires more hidden layers to fit
cx1,cy1,r1=0.3,0.3,0.3
cx2,cy2,r2=-0.4,-0.4,0.3

# create the data
X=2*(rand(ns,2) .-0.5)
C1=(X[:,1] .-cx1).^2 .+ (X[:,2] .-cy1).^2 .<r1^2
C2=(X[:,1] .-cx2).^2 .+ (X[:,2] .-cy2).^2 .<r2^2
T=1.0*(C1.|C2)

# do the plots
figure(figsize=(4,4))
for k=1:ns
    T[k]==1 ? plot(X[k,1],X[k,2],"go",markersize=3) : ()
    T[k]==0 ? plot(X[k,1],X[k,2],"r+",markersize=3) : ()
end;
```

```julia
# bagel alternative!

ns=200

# two circular blobs - requires more hidden layers to fit
cx1,cy1,r1=0.0,0.0,0.7
cx2,cy2,r2=0.0,0.0,0.3

# create the data
X=2*(rand(ns,2) .-0.5)
C1=(X[:,1] .-cx1).^2 + (X[:,2] .-cy1).^2 .<r1^2
C2=(X[:,1] .-cx2).^2 + (X[:,2] .-cy2).^2 .<r2^2

T=zeros(ns)

for k=1:ns
   if C1[k]&(~C2[k]).==1
    T[k]=1
    end
end

# do the plots
figure(figsize=(4,4))
for k=1:ns
    T[k]==1 ? plot(X[k,1],X[k,2],"go",markersize=3) : ()
    T[k]==0 ? plot(X[k,1],X[k,2],"r+",markersize=3) : ()
end;
```

### 8.2.2 **<font color='red'>Question.</font>** Functions for the Prediction and Cost

* The input-to-output transformations are   

$$ f(\tilde{\mathbf{X}}\mathbf{w})=\mathbf{H}~~\mbox{ and} ~~f(\tilde{\mathbf{H}}\mathbf{v})=\mathbf{P}$$ 


$\hspace{0.6cm}$ where $f(z)=1/(1+e^{-z})$.


* The cost function is   


$$C=-\frac{1}{n_s}\sum_s\left[T_s\log(P_s)+(1-T_s)\log(1-P_s)\right]$$


* Create two functions that take $\tilde{\mathbf{X}}$, $\mathbf{w}$, $\mathbf{v}$ and $\mathbf{T}$ as inputs (as required) and then output:   
(i) the prediction $\mathbf{P}$ and hidden layer $\mathbf{\tilde{H}}$ values and   
(ii) the cost $C$.

### 8.2.2 **<font color='green'>Answer.</font>** Functions for the Prediction and Cost


* The input-to-output transformations are   

$$ f(\tilde{\mathbf{X}}\mathbf{w})=\mathbf{H}~~\mbox{ and} ~~f(\tilde{\mathbf{H}}\mathbf{v})=\mathbf{P}$$ 


$\hspace{0.6cm}$ where $f(z)=1/(1+e^{-z})$.


* The cost function is   


$$C=-\frac{1}{n_s}\sum_s\left[T_s\log(P_s)+(1-T_s)\log(1-P_s)\right]$$



```julia
# logistic function
f(z)=1 ./(1 .+exp.(-z))

# also returns the state of the hidden layer
# this is useful later for backprop
function MyPrediction(Xb,w,v)
    ns=size(Xb,1)
    Hb=[f(Xb*w) ones(ns)]
    P=f(Hb*v)
    return P,Hb
end

function MyCost(Xb,w,v,T)
    ns=length(T)
    P,~=MyPrediction(Xb,w,v)
    C=-(1/ns)*sum(T.*log.(P) .+(1 .-T).*log.(1 .-P))
end
```

### 8.2.3 **<font color='red'>Question.</font>** Cost Function Gradients

* The following relations are required for the weight updates


$$
\frac{dC}{d\mathbf{v}}=\frac{1}{n_s}\tilde{\mathbf{H}}'\Delta^p  ~~\mbox{ and }~~\frac{dC}{d\mathbf{w}}=\frac{1}{n_s}\tilde{\mathbf{X}}'\Delta^h
$$


* Write a function that takes $\tilde{\mathbf{X}}$, $\tilde{\mathbf{H}}$, $\mathbf{w}$, $\mathbf{v}$ and $\mathbf{T}$ and outputs the gradients $\frac{dC}{d\mathbf{v}}$ and $\frac{dC}{d\mathbf{w}} $.

### 8.2.3 **<font color='green'>Answer.</font>** Cost Function Gradients

```julia
function MydCdwdCdv(Xb,Hb,v,P,T)
    ns=size(Xb,1)
    
    # update rule for the v weights
    Deltap=(P-T)
    dCdv=((Hb')*(Deltap))/ns
    
    # update rule for the w weights
    M=Deltap*(v[1:end-1]') #' 
    H=Hb[:,1:end-1];
    Deltah=H.*(1 .-H).*M
    dCdw=((Xb')*(Deltah))/ns
        
    return dCdw,dCdv

end
;
```

### 8.2.4 **<font color='red'>Question.</font>** Gradient Descent



* We now have all the functions required to iterate the gradient descent. 



* Write a final function that contains a loop over $n_u$ updates of the weights.   
Make $n_h$, the size of the hidden layer, and argument of the function.   




$$ \mathbf{w}_\mathrm{new}=\mathbf{w}_\mathrm{old}-\alpha\frac{dC}{d\mathbf{w}} 
~~~\mbox{ and }~~~
\mathbf{v}_\mathrm{new}=\mathbf{v}_\mathrm{old}-\alpha\frac{dC}{d\mathbf{v}}$$ 


* $\alpha$ is a positive learning rate that you can choose.  


* Plot a graph of the cost $C$ as a function of the iteration steps and check that it is monotonically decreasing.




* For $n_h=2,4,6,8$ what is the percentage of correctly classified data points?

### 8.2.4 **<font color='green'>Answer.</font>** Gradient Descent

```julia
# This is the loop over weight updates
function MyGradientDescent(X,T,nh,a,nu)
    
    ns=size(X,1)
    nx=size(X,2)
    Xb=[X ones(ns,1)]
    Hb=zeros(ns,nh+1)
    
    wnew,wold=randn(nx+1,nh),zeros(nx+1,nh)
    vnew,vold=randn(nh+1,1),zeros(nh+1,1)
    
    su=collect(1:nu)
    Cu=zeros(nu,1)
    P=zeros(ns,1)
    
    for i=1:nu
        wold=wnew 
        vold=vnew
        
        P,Hb=MyPrediction(Xb,wold,vold)
        Cu[i]=MyCost(Xb,wold,vold,T)
        
        dCdw,dCdv=MydCdwdCdv(Xb,Hb,vold,P,T)
        
        wnew=wold-a*dCdw
        vnew=vold-a*dCdv
    end

    H=Hb[:,nh]
    
    return su,Cu,wold,vold,H,P
    
end

;
```

```julia
# now run the program

nh=10
a=2.0
nu=10000
su,Cu,w,v,H,P=MyGradientDescent(X,T,nh,a,nu);

figure(figsize=(8,5))

subplot(221)
plot(su,Cu)
xlabel("iterations")
ylabel("cost")

subplot(222)
semilogx(su,Cu)
xlabel("iterations")
ylabel("cost");

tight_layout()
```

### 8.2.5 Plotting the Classifier and Hidden Layers

*  This data is not linearly separable;   
however, you can interpret the weights between the input and hidden layer as defining $n_h$ lines   
(like was done for the simple input-output case). 


* Plot the $n_h$ lines represented by the weight matrix $\mathbf{w}$ to gain insight into the role of the hidden layer.


* Plot a grid of points on the plane that are colour coded by the predictions. 


* It is useful to write functions that will do the plotting for you.  


```julia
function MyPlotLine(w,xl,xh,S)
    m=-w[1]/w[2]
    c=-w[3]/w[2]
    plot([xl,xh],m*[xl,xh] .+c,S)
end

function MyScatterPlot(xr,yr,dx,dy,w,v)
    xx=collect(-xr:dx:xr); nxx=length(xx)
    yy=collect(-yr:dy:yr); nyy=length(yy)
    c=0
    Xb=zeros(nxx*nyy,3)
    for i=1:nxx
        for j=1:nyy
            c=c+1
            Xb[c,1]=xx[i]
            Xb[c,2]=yy[j]
            Xb[c,3]=1.0
        end
    end
    P,~=MyPrediction(Xb,w,v)
    s0=findall(P[:].<0.5)
    s1=findall(P[:].>0.5)        
    plot(Xb[s0,1],Xb[s0,2],"ro",markersize=2)
    plot(Xb[s1,1],Xb[s1,2],"go",markersize=2)
end
;    
```

```julia
figure(figsize=(8,3.5))

# do the plots
subplot(121)
for k=1:ns
    T[k]==1 ? plot(X[k,1],X[k,2],"go",markersize=4) : ()
    T[k]==0 ? plot(X[k,1],X[k,2],"ro",markersize=4) : ()
end

for i=1:nh
    MyPlotLine(w[:,i],-1,1,"k")
end
axis([-1,1,-1,1])
title("Data with hidden-layer lines")

subplot(122)
MyScatterPlot(1,1,0.01,0.01,w,v);
axis([-1,1,-1,1]);


for i=1:nh
    MyPlotLine(w[:,i],-1,1,"k")
end
axis([-1,1,-1,1])
title("Prediction with hidden-layer lines")
;
```
